```python
# Index tunggal, field index
import numpy as np
a = np.arange(10)
print(a)
b = a[5]
print(b)
```

```python
a = np.arange(10)
print(a)
s = slice(2,7,2)
print(a[s])
```

```python
a = np.arange(10)
b = a[2:7:2]
print(b)
```

```python
a = np.arange(10)
print(a)
print(a[2:])
```

```python
a = np.arange(10)
print(a)
print(a[2:5])
```

```python
a = np.array([[1,2,3],[3,4,5],[4,5,6]])
print(a)

print('Slice array dari index')
print(a[1:])
```

```python
a = np.array([[1,2,3],
             [3,4,5],
             [4,5,6]])

print('Array kita:')
print(a)
print('\n')

print('item dari kolom kedua:')
print(a[...,1])
print('\n')

print('item dari baris kedua:')
print(a[1,...])
print('\n')

print('item dari kolom 1 dan seterusnya:')
print(a[...,1:])
```
